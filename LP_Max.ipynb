```python
import gurobipy as gp
```

image.png

```python
# Criando o modelo

modelo = gp.Model()
```

```python
# Adicinar variavéis de decisão

x1 = modelo.addVar()
x2 = modelo.addVar()
x3 = modelo.addVar()
```

```python
# Função objetivo

modelo.setObjective(12 * x1 + 18 * x2 + 22 * x3, sense=gp.GRB.MAXIMIZE)
```

```python
# Sujeito as seguintes restrições

c1 = modelo.addConstr(1.5 * x1 + 1.2 * x3 <= 120)

c2 = modelo.addConstr(2.2 * x2 + 1.4 * x3 <= 220)

c3 = modelo.addConstr(1.2 * x1 + 2 * x2 + 2.4 * x3 <= 250)

c4 = modelo.addConstr(x1 + x2 + x3 >= 0)
```

```python
# Resolver o modelo

modelo.optimize()
```

```python
# mix de produção

print("Produto A:",x1.X)
print("Produto B:",x2.X)
print("Produto C:",x3.X)
```

```python
# Ocupação das maquinas

tempo_m1 = 120 - c1.Slack
print("Maquina 1:", tempo_m1)

tempo_m2 = 200 - c2.Slack
print("Maquina 2:", tempo_m2)

tempo_m3 = 250 - c3.Slack
print("Maquina 3:", tempo_m3)
```
